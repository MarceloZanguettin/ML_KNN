Objetivo: Sistema de Recomendação de filme com K-NN

```python
# Importar as bibliotecas necessárias
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

```python
# Carregar Dados
df = pd.read_csv("/content/drive/MyDrive/filmes.csv", index_col=0)
df
```

```python
# Remover a coluna label e Movie Name
df = df.drop("Label", axis=1)
df = df.drop("Movie Name", axis=1)
df
```

```python
# Preparar o algortimo
knn = NearestNeighbors(n_neighbors=3)
knn.fit(df)
```

```python
# Gostei do filme 6 Black Panther
knn.kneighbors([[8.8,0,1,0,0,0,0,0]])
```
